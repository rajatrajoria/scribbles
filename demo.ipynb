```python
def is_contour_darker(contour, image = "./test4.jpg"):
    # Create a mask for the contour
    contour_mask = np.zeros_like(image, dtype=np.uint8)
    cv2.drawContours(contour_mask, [contour], -1, (255, 255, 255), cv2.FILLED)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Calculate the average intensity value within the contour
    contour_pixels = gray[np.where(contour_mask[:, :, 0] == 255)]
    contour_avg_intensity = np.mean(contour_pixels)

    # Calculate the average intensity value of the entire image
    image_avg_intensity = np.mean(gray)

    # Compare the average intensity values
    if contour_avg_intensity < image_avg_intensity:
        plt.
        print("T")
        return True
    else:
        return False
```

```python
import cv2
import numpy as np

def draw_rectangles(image_path = "./test4.jpg"):
    # Load the image
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply image thresholding to obtain binary image
    ret, threshold = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Find contours in the thresholded image
    contours, _ = cv2.findContours(threshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Iterate through the contours and draw rectangles around them
    for contour in contours:
        if(is_contour_darker(contour, image)):
            x, y, w, h = cv2.boundingRect(contour)
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 1)

    # Display the image with rectangles
    cv2.imshow('Rectangles', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

```
